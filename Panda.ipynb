```python
import pandas as pd
```

```python
#series 
data=[10,20,30]  #list

series=pd.Series(data,index=["a","b","c"])  # .Series()- is a constructor 
print(series)
series.loc["c"]=40   # iloc[0] - for integer location 
print(series.loc["c"])

print(series[ series<30])

cal={"D1":1900,"D2":1850,"D3":1800}  # making a dictionary
series2=pd.Series(cal)
print(series2.loc["D3"])
```

```python
#dataframe - A tabular structure 
data={ "name":["bob","rob","ned","Jon"] ,
        "age":[30,21,50,20]}

dataframe=pd.DataFrame(data,index=["char1","char2","char3","char4"])
# add coloum 
dataframe["job"]=["cook","prince","king","night-watch"]
new_row= pd.DataFrame([{"name":"strak","age":"N\D","job":"north"}])
dataframe=pd.concat([dataframe,new_row])

print(dataframe)
print(dataframe.loc["char2"])
print(dataframe.iloc[2])
```

```python
#filtering
tall=df[df["Height"]>=2]
legend= df[df["Legendary"]== True ]
water= df[ (df["Type1"]=="Water") | (df["Type2"]=="Water")] # | for OR  & - AND
print(water)
```

```python
#agrigate function 
print(df.mean(numeric_only = True))   #.sum , .min , max , .count
print(df.count())
print(df["Height"].sum())

group = df.groupby("Type1")
print(group["Height"].sum())
```

```python
# cleaning
#drop coloum 
dfL= df.drop(columns=["Legendary"])
print(dfL)
#missing  dropna= drop not avalible 
dfT = df.dropna(subset=["Type2"])
dft = df.fillna({"Type2":"---"})
print(dfT.count())
print(dft)
df["Type1"]= df["Type1"].replace({"Grass":"GRASS",
                                  "Fire":"FIRE"})
df["Name"]= df["Name"].str.upper()
df["Legendary"]=df["Legendary"].astype(bool)   ## to fix data type
df=df.drop_duplicates()
print(df)
```
